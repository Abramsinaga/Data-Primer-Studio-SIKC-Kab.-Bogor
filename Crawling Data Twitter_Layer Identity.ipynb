```python
# Crawl Data

filename = 'parung.csv'
search_keyword = 'parung -aceh -amin -caleg -anies -anis -ganjar -mahfud -pilpres -pemilu -prabowo -gibran -amin -imin lang:id since:2020-01-01 until:2023-12-31'
limit = 1500

!npx --yes tweet-harvest@2.2.8 -o "{filename}" -s "{search_keyword}" -l {limit} --token {twitter_auth_token}
```

```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=";")

# Display the DataFrame
display(df)
```
