```python
import pandas as pd

# Membaca file CSV dan memilih hanya kolom "full_text"
def pilih_kolom_full_text(input_file, output_file):
    # Membaca file CSV
    df = pd.read_csv(file_path, delimiter=",")

    # Memilih hanya kolom "full_text"
    if 'full_text' in df.columns:
        selected_column = df['full_text']
    else:
        print("Kolom 'full_text' tidak ditemukan.")
        return

    # Menyimpan hasil pemilihan ke file teks
    selected_column.to_csv(output_file, index=False)

filepathoutput =  f"{folder_name}{file_name}.csv"  #ISI INI JUGAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAA
#  pemanggilan fungsi dengan file input dan output
pilih_kolom_full_text(file_path, filepathoutput)
```

```python
import re

def bersihkan_teks(teks):
    teks = teks.lower() #membuat menjadi huruf kecil
    # Menghapus simbol dan ikon percakapan
    teks_bersih = re.sub(r'@[A-Za-z0-9_]+', '', teks)  # Menghapus mention
    teks_bersih = re.sub(r'#', '', teks_bersih)  # Menghapus tanda pagar
    teks_bersih = re.sub(r'RT[\s]+', '', teks_bersih)  # Menghapus retweet
    teks_bersih = re.sub(r'https?://\S+', '', teks_bersih)  # Menghapus tautan
    teks_bersih = re.sub(r'&amp;', '', teks_bersih)  # Menghapus simbol HTML untuk '&'
    teks_bersih = re.sub(r'[^\x00-\x7F]+', '', teks_bersih)  # Menghapus karakter non-ASCII
    teks_bersih = re.sub(r'\n', '', teks_bersih)  # Menghapus karakter newline

    return teks_bersih
```

```python
# Membaca teks dari file input
nama_file_input = filepathoutput
nama_file_output =  f"{folder_name}{file_name}_bersih.csv"  #ISI INI JUGAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAA

with open(nama_file_input, 'r', encoding='utf-8') as file_input:
    teks_twitter = file_input.read()

# Membersihkan teks
teks_bersih = bersihkan_teks(teks_twitter)

# Menulis teks bersih ke file output
with open(nama_file_output, 'w', encoding='utf-8') as file_output:
    file_output.write(teks_bersih)

print("Data berhasil dibersihkan dan disimpan di", nama_file_output)
```

```python
# Membaca stopword dari file teks
def baca_stopword(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        stopwords = file.read().splitlines()
    return stopwords

# Path file stopword
file_stopword = '/content/drive/MyDrive/id_stopwords.txt'

# Menggunakan fungsi untuk membaca stopword
stopwordsid = baca_stopword(file_stopword)


```

```python
#tidak perlu dieksekusi kalo tidak perlu
# Nama file teks
nama_file = file_stopword

# Buka file teks dan baca isinya
with open(nama_file, 'r', encoding='utf-8') as file:
    isi_file = file.readlines()

# Tambahkan beberapa kata satu per satu
kata_baru = ['dari', 'dan']  # Ganti dengan kata-kata yang ingin di tambahkan

# Menambahkan kata baru satu per satu menggunakan append
for kata in kata_baru:
    isi_file.append('\n'+kata)

# Simpan kembali isi file dengan kata-kata baru
with open(nama_file, 'w', encoding='utf-8') as file:
    file.writelines(isi_file)

print("Kata-kata baru telah ditambahkan ke file:", nama_file)
```
